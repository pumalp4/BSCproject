```python
import numpy as np
from matplotlib.pyplot import figure, show, xlabel, ylabel, legend, ylim, xlim, title, yscale, xscale, errorbar
```

```python
def Compare(A,B,Title='Undefined Title',name1='',name2=''):
    # Function to compare LIFEsim results to a blackbody
    # A is the path to result file of chosen spectrum
    # B is the path to the blackbody result found using LIFEsim
    # Title is the desired title of the graph
    
    Data = np.loadtxt("LIFEsim FIles/{}".format(A),delimiter=' ' )
    Data2 = np.loadtxt("LIFEsim FIles/{}".format(B), delimiter=' ')
    
    #print(Data,BB)
    
    
    lam_Data = Data[:,0]*10**(6)
    
    SNR_Data = Data[:,1]
    
    F_Data = Data[:,2]
    sigma_Data = F_Data/SNR_Data
    print(sigma_Data)
    
    lam_Data2 = Data2[:,0]*10**(6)
    SNR_Data2 = Data2[:,1]
    F_Data2 = Data2[:,2]
    sigma_Data2 = F_Data2/SNR_Data2
    #print(sigma_BB)
    
    
    fig = figure()
    fr = fig.add_subplot(1,1,1)
    fr.plot(lam_Data,F_Data,color='r',label='{} Flux'.format(name1))
    fr.fill_between(lam_Data, F_Data-sigma_Data, F_Data+sigma_Data)
    #fr.fill_between(lam_Data, F_Data-sigma_Data, F_Data+sigma_Data)
    #fr.plot(lam_Data,SNR_Data,color='k',label='{} SNR'.format(Title))
    #fr.errorbar(lam_Data,F_Data,yerr=sigma_Data, label='Ultramafic Flux')
    
    
    fr.plot(lam_Data2,F_Data2,color='k',label=name2)
    #fr.plot(lam_,SNR_,color='c',label='Blackbody SNR')
    #fr.errorbar(lam_Data2,F_Data2, yerr=sigma_Data2, label=name2)
    fr.fill_between(lam_Data2, F_Data2-sigma_Data2, F_Data2+sigma_Data2,color='g')
    ylabel('Signal per bin [$e^{-}s^{-1}bin^{-1}$]')
    xlabel('Wavelength [microns]')
    
    yscale('log')
    title(Title)
    legend()
    
    show()
    
    
    fig1 = figure()
    fr1 = fig1.add_subplot(1,1,1)
    fr1.plot(lam_Data,F_Data,color='r',label='{} Flux'.format(name1))
    #fr1.fill_between(lam_Data, F_Data-sigma_Data, F_Data+sigma_Data)
    #fr.plot(lam_Data,SNR_Data,color='k',label='{} SNR'.format(Title))
    #fr.errorbar(lam_Data,F_Data,yerr=sigma_Data, label='Ultramafic Flux')
    
    
    fr1.plot(lam_Data2,F_Data2,color='m',label=name2)
    #fr1.fill_between(lam_Data2, F_Data2-sigma_Data2, F_Data2+sigma_Data2)
    #fr.plot(lam_BB,SNR_BB,color='c',label='Blackbody SNR')
    #fr.errorbar(lam_BB,F_BB, yerr=sigma_BB, label='Blackcody Flux')
    ylabel('Signal per bin [$e^{-}s^{-1}bin^{-1}$]')
    xlabel('Wavelength [microns]')
    
    yscale('log')
    title(Title)
    legend()
    
    show()
    
def Graph(file,error='no'):
    
    
    Data = np.loadtxt("LIFEsim FIles/{}".format(file),delimiter=' ' )
    lam = Data[:,0]*10**(6)
    SNR = Data[:,1]
    F = Data[:,2]
    
    sigma = F/SNR
    
    fig = figure()
    fr = fig.add_subplot(1,1,1)
    fr.plot(lam,F,label='Flux')
    if error=='yes':
        fr.fill_between(lam, F-sigma,F+sigma,color='g')
    
    xlabel('Wavelenght [microns]')
    ylabel('Flux [photon micron-1 m-2 s-1]')
    title('Flux')
    show()
    
    fig2 = figure()
    fr2 = fig2.add_subplot(1,1,1)
    fr2.plot(lam,SNR,label='SNR')
    xlabel('Wavelenght [microns]')
    ylabel('Flux [photon micron-1 m-2 s-1]')
    title('SNR')
    show()
```

```python
Compare('Granitoid.txt','Blackbody_earthlike.txt','Earthlike Granitoid')
```

```python
Compare('Metal-Rich.txt','Blackbody_earthlike.txt','Earthlike Metal-Rich')
```

```python
Compare('Metal-Rich.txt','Granitoid.txt','Earthlike Metal-Rich',name1='Metal-Rich',name2='Granitoid')
```

```python
Graph('Granitoid.txt',error='yes')
Graph('Metal-Rich.txt',error='es')
```
