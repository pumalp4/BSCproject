```python
import numpy as np
from matplotlib.pyplot import figure, show, xlabel, ylabel, legend, ylim, xlim, title
```

```python
MR = np.loadtxt("Graph 0.15 AU/Metal-Rich_015.csv",delimiter=',')
UM = np.loadtxt("Graph 0.15 AU/Ultramafic_015.csv",delimiter=',')
FP = np.loadtxt("Graph 0.15 AU/Feldspathic_015.csv",delimiter=',')
B = np.loadtxt("Graph 0.15 AU/Basaltic_015.csv",delimiter=',')
Gr = np.loadtxt("Graph 0.15 AU/Granitoid_015.csv",delimiter=',')
Cl =np.loadtxt("Graph 0.15 AU/Clay_015.csv",delimiter=',')
Fe =np.loadtxt("Graph 0.15 AU/Fe-Oxidised_015.csv",delimiter=',')



fig = figure()
fr = fig.add_subplot(1,1,1)
fr.plot(MR[:,0], MR[:,1],color='r',label='Metal-Rich')
fr.plot(UM[:,0], UM[:,1],color='b',label='Ultramafic')
fr.plot(FP[:,0], FP[:,1],color='limegreen',label='Feldspathic')
fr.plot(B[:,0], B[:,1],color='m',label='Basaltic')
fr.plot(Gr[:,0], Gr[:,1],color='c',label='Granitoid')
fr.plot(Cl[:,0], Cl[:,1],color='peru',label='Clay')
fr.plot(Fe[:,0], Fe[:,1],color='grey',label='Fe-Oxidised')
xlabel("$\lambda$ [microns]")
ylabel("Brightness Temperature [K]")
ylim((500,1200))
xlim((3,25))
title("0.15AU")
legend(loc='upper right')
show()


```
